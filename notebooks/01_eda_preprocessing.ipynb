```python
print("=== IQR 기반 수치형 특성별 이상치 탐지 ===")
print("-" * 50)

for col in numeric_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # 상한선과 하한선 계산 (일반적으로 1.5배수 적용)
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # 상한선 또는 하한선을 벗어나는 조건 필터링
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
    print(f"[{col}]")
    print(f"  - IQR 범위: {Q1} ~ {Q3} (IQR = {IQR:.2f})")
    print(f"  - 정상 범위 기준: {lower_bound:.2f} ~ {upper_bound:.2f}")
    print(f"  - 탐지된 이상치 개수: {len(outliers)}개")
    if len(outliers) > 0:
        print(f"  - 실제 이상치 값들: {outliers.values}")
    print("-" * 50)

    # 이상치 수학적 탐지(IQR)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, OneHotEncoder

def clean_and_prepare_data(df_input):
    """
    파이프라인 진입 전, 중복 제거 및 불필요한 컬럼을 정제하는 함수
    """
    df_clean = df_input.copy()
    
    # 1. 중복 레코드 제거
    duplicated_count = df_clean.duplicated().sum()
    if duplicated_count > 0:
        print(f"[정제] 중복 데이터 {duplicated_count}건을 제거합니다.")
        df_clean = df_clean.drop_duplicates()
        
    # 2. 완전히 비어있는(모든 값이 결측치인) 컬럼 제거
    empty_cols = [col for col in df_clean.columns if df_clean[col].isnull().all()]
    if empty_cols:
        print(f"[정제] 완전히 비어있는 컬럼 제거: {empty_cols}")
        df_clean = df_clean.drop(columns=empty_cols)
        
    return df_clean

# --- [단계 1] 원본 데이터 병합 및 정제 ---
X = heart_disease.data.features 
y = heart_disease.data.targets 
raw_df = pd.concat([X, y], axis=1)

# 타겟 이진 분류화 및 데이터 정제
raw_df['target'] = raw_df['num'].apply(lambda x: 1 if x > 0 else 0)
raw_df = raw_df.drop(columns=['num'])

cleaned_df = clean_and_prepare_data(raw_df)

# --- [단계 2] Feature와 Target 분리 및 Train/Test 분할 ---
X_clean = cleaned_df.drop(columns=['target'])
y_clean = cleaned_df['target']

# 클래스 비율을 유지하며 분할 (stratify 적용)
X_train, X_test, y_train, y_test = train_test_split(
    X_clean, y_clean, test_size=0.2, random_state=42, stratify=y_clean
)

# --- [단계 3] 데이터 타입별 타겟 컬럼 정의 ---
# 앞서 분석한 기술 통계량을 바탕으로 정의합니다.
numeric_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# --- [단계 4] 컴포넌트별 파이프라인 구성 (sklearn.pipeline.Pipeline) ---
# 1) 연속형 변수 파이프라인: 중앙값 대치 -> Robust 스케일러 적용
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())
])

# 2) 범주형 변수 파이프라인: 최빈값 대치 -> 원핫 인코딩(학습 때 없던 카테고리는 무시)
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# 3) ColumnTransformer를 이용해 컬럼별로 파이프라인을 매핑
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# --- [단계 5] 파이프라인 실행 및 데이터 누수 검증 ---
# Train 데이터로만 기준을 잡습니다 (fit_transform)
X_train_transformed = preprocessor.fit_transform(X_train)

# Test 데이터는 이미 학습된 기준을 '적용'만 합니다 (transform) - 데이터 누수 원천 차단!
X_test_transformed = preprocessor.transform(X_test)

print("\n=== 파이프라인 구성 및 변환 완료 ===")
print(f"변환된 훈련 데이터 크기(Shape): {X_train_transformed.shape}")
print(f"변환된 테스트 데이터 크기(Shape): {X_test_transformed.shape}")

# Scikit-Learn 기반 프로덕션 파이프라인 구현
```

**EDA를 통해 발견한 데이터 특징**

1. 타겟 클래스의 비율 (Class Distribution)
   - 심장병 환자(1)와 정상군(0)의 비율이 약 46% 대 54%로 확인되었습니다. 완전한 극단적 불균형은 아니지만 약간의 불균형이 존재합니다. 모델 평가 시 단순 '정확도(Accuracy)'만 보면 성능이 과대평가될 위험이 있습니다.

2. 연속형 변수의 이상치 (Outliers in Numeric Features)
   - Boxplot과 IQR 탐지 결과, `chol`(콜레스테롤)과 `trestbps`(휴식기 혈압) 등에서 정상 범위를 크게 벗어나는 극단적인 고수치 이상치들이 발견되었습니다.이 이상치들은 제거해야 할 오류 데이터가 아니라, 실제 심장병 고위험군 환자의 중요한 임상 신호(Signal)입니다.

3. 결측치의 존재 (Missing Values)
   - 수치형 변수에는 결측치가 없으나, 범주형/정수형 특성인 `ca`와 `thal` 변수에서 일부 결측치가 존재함을 확인했습니다.
